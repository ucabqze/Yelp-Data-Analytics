```python
import pandas as pd
import numpy as np
import json
import gc
from tqdm import tqdm_notebook as tqdm
```

```python
attributes = pd.read_csv('processed_attributes_round13.csv')
```

```python
attributes = attributes.drop('Unnamed: 0', axis=1)
```

```python
tags = list(attributes.columns)
```

```python
dropable_instance = set()
#tags = ['address', 'attributes', 'business_id', 'categories', 'city', 'hours', 'is_open', 'latitude', 'longitude', 'name', 'postal_code', 'review_count', 'stars', 'state']
#business = pd.read_csv('yelp_academic_dataset_business.csv', low_memory=False)
for i in tqdm(range(len(attributes))):
    for tag in tags:
        if pd.isnull(attributes.at[i,tag]):
            dropable_instance.add(i)

attributes.drop(attributes.index[list(dropable_instance)], inplace=True)
print('Successfully drop %d instances'% len(dropable_instance))
```

```python
attributes.to_csv('attributes_no_null_13.csv', encoding='utf-8', index=False)
```
